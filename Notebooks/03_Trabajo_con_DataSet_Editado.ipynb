```python
import pandas as pd

df=pd.read_csv('../data/Video-Games-Sales-Editado.csv')

df.head()
```

```python
df.groupby(by='Genre').agg({
    'Global':['sum','min','max','mean']
}).sort_values(by=[('Global', 'sum')], ascending=False)
```

```python
juegos_review_alto = df.sort_values(by="Review",ascending=False)
juegos_review_alto.head()
```

```python
generos_top = df["Genre"].value_counts().sort_values(ascending=False)
generos_top
```

```python
df.corr(numeric_only=True,method='pearson')
```

Se observa una correlación alta entre las ventas Globales y las de Norte America.

```python
df_Wii=df[df['Platform'] =='Wii']
df_Wii.corr(numeric_only=True,method='pearson')
```

```python
preferencias_regionales_por_genero = df.groupby('Genre').agg({
    'North America': 'sum',
    'Europe': 'sum',
    'Japan': 'sum',
    'Rest of World': 'sum'
})

# Calculating the total sales in each region for normalization
total_ventas_por_region = preferencias_regionales_por_genero.sum()

# Normalizing the sales data to understand relative popularity of genres in each region
preferencia_normalizada_por_genero = preferencias_regionales_por_genero / total_ventas_por_region

#print(total_ventas_por_region)

#print(preferencias_regionales_por_genero)

#print(preferencia_normalizada_por_genero)
```

```python
publisher_ventas_review = df.groupby('Publisher').agg({'Global': 'sum', 'Review': 'mean'}).reset_index()

top_publisher_ventas = publisher_ventas_review.sort_values(by='Global', ascending=False).head(10)

print(top_publisher_ventas)
```

# Analizaremos al Publisher Nintendo

```python
df_nintendo=df[df['Publisher'] =='Nintendo']
juegos_top_review = df_nintendo.sort_values(by='Review', ascending=False)
juegos_top_review.head(10)
```

Observamos que la saga The Legend of Zelda posee muy buenos reviews y ventas para Nintendo.

```python
juegos_top_ventas = df_nintendo.sort_values(by='Global', ascending=False)
juegos_top_ventas.head(10)
```

```python
top_ventas_wii=juegos_top_ventas[juegos_top_ventas['Platform'] =='Wii']
top_ventas_wii.head()
```

Notamos que juegos de la plataforma Wii lograron que las ventas globales de Nintendo fueran significativas en comparación con los demás Publishers.
