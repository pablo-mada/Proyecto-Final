# Descripción de las columnas:  
- `index`: Indice
- `Rank`: El Ranking de acuerdo a las ventas globales.    
- `Game Title`: Titulo del juego.  
- `Platform`: Plataforma en la cual el juego está disponible.  
- `Year`: Año de lanzamiento.  
- `Genre`: Genero del juego.  
- `Publisher`: La compañia que publica el juego.  
- `North America`: El número de ventas en Norte America, en millones.  
- `Europe`: El número de ventas en Europa, en millones.  
- `Japan`: El número de ventas en Japon, en millones.  
- `Rest of World`: El número de ventas en el resto del mundo, en millones.  
- `Global`: El número de ventas Globales, en millones.  
- `Review`: El puntaje de juego de acuerdo a sus reviews.
